```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from matplotlib_venn import venn3, venn3_circles
```

```python
plt.subplots_adjust()
fig, ax = plt.subplots(figsize=(6,6))
#fig.set_linewidth(1.0)
#fig.set_edgecolor("black")
#fig.set_facecolor('#d3d3d3')
#fig.set_linewidth(figsize=(4,4), edgecolor="black", linewidth=1.0)
ax.set_title("Venn Diagram", fontsize=16)
ax.set_title("U=23", loc="left", size="medium")
ax.axis('off')
border = patches.Rectangle((0, 0.009), 0.99, 0.99, 
                          linewidth=2, 
                          edgecolor='black', 
                          facecolor='#d3d3d3', 
                          transform=ax.transAxes)
ax.add_patch(border)
ax.text(0.8,0.18,'y, z',fontsize=12)

ax2 = fig.add_axes((0.11, 0.1, 0.8, 0.8))

set_a = {'a', 'b', 'c', 'k', 'm', 'n', 'p', 'q', 'r', 'v', 'w', 'x'}
set_b = {'d', 'e', 'f', 'k', 'm', 'n', 's', 't', 'u', 'v', 'w', 'x'}
set_c = {'g', 'h', 'j', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x'}

label_map = {
    '100': 'a, b, c',
    '010': 'd, e, f',
    '001': 'g, h, j',
    '110': 'k, m, n',
    '101': 'p, q, r',
    '011': 's, t, u',
    '111': 'v, w, x'
}

v = venn3([set_a, set_b, set_c], set_labels=('A', 'B', 'C'))
c = venn3_circles([set_a, set_b, set_c])

for subset_id, text in label_map.items():
    label = v.get_label_by_id(subset_id)
    patch = v.get_patch_by_id(subset_id)
    if label:
        label.set_text(text)
    if subset_id == '111':
        patch.set_color('white')

plt.show()
```
